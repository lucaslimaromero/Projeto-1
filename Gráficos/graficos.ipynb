## Complexidade de Tempo
- A seguir estão os gráficos referentes à complexidade de tempo dos algoritmos de ordenação
- Formato: Tempo (t) x Tamanho do Vetor de Entrada (n)

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Bubble Sort Normal
- O(n²)

```python
# Criando Eixo X
x = [1000, 11000, 21000, 31000, 41000, 51000, 61000]

# Criando Eixo y
y = [0.0036, 0.5416, 1.9965, 5.3039, 8.9061, 12.9281, 18.1517]

# Plotando o gráfico
plt.scatter(x, y)
plt.xlabel('Tamanho do Vetor')
plt.ylabel('Tempo Médio (s)')
plt.grid()
plt.show()

plt.scatter(x, y)
plt.xlabel('Tamanho do Vetor')
plt.ylabel('Tempo Médio (s)')
plt.grid()
plt.loglog()
plt.show()
```

### Bubble Sort Aprimorado
- O(n²)

```python
# Criando Eixo X
x = [1000, 11000, 21000, 31000, 41000, 51000, 61000]

# Criando Eixo y
y = [0.0022, 0.3301, 1.4094, 3.3463, 5.1272, 8.2471, 11.8324]

# Plotando o gráfico
plt.scatter(x, y, color='red')
plt.xlabel('Tamanho do Vetor')
plt.ylabel('Tempo Médio (s)')
plt.grid()
plt.show()

plt.scatter(x, y, color='red')
plt.xlabel('Tamanho do Vetor')
plt.ylabel('Tempo Médio (s)')
plt.grid()
plt.loglog()
plt.show()
```
